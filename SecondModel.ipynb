###Importing Everthing

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import accuracy_score, mean_absolute_error, mean_squared_error, r2_score, confusion_matrix, classification_report
```

```python
df=pd.read_csv("Dataset .csv")
```

```python
df
```

```python
df.isna().sum()
```

```python
df.fillna(method='ffill', inplace=True)
```

```python
df.info()
```

```python
df[df==0].count()
```

```python
df['Cuisines']=df['Cuisines'].str.lower()
```

```python
df
```

```python
#Determining the criteria for restaurant recommendations
user_cuisine_preference = 'north indian'
user_price_range_preference = 2.0 #assuming 2.0 correspods to moderate-priced restaurants
```

```python
#Implement Content based filtering using Decision Tree Regressor
#Encoding "Cuisines" using one-hot encoding
x=pd.get_dummies(df['Cuisines'],drop_first=True)
y=df['Aggregate rating']
```

```python
x
```

```python
y
```

##Test/Train Split


```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
y_train
```

###Using Decision Tree

```python
model = DecisionTreeRegressor()
model.fit(x_train,y_train)
```

```python
user_cuisine_encoded = pd.get_dummies([user_cuisine_preference.lower()],prefix='',prefix_sep='').reindex(columns=x.columns,fill_value=0)
```

```python
predicted_rating=model.predict(user_cuisine_encoded)
```

```python
print(f"User Preference: Cuisine={user_cuisine_preference},Price Range={user_price_range_preference}")
print(f"Predicted Restaurant Rating: {predicted_rating[0]}")
```

##Errors

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
y_pred=model.predict(x_test)
#mean_squared_error
print("MSE:",mean_squared_error(y_test,y_pred))
#mean_absolute_error
print("MAE:",mean_absolute_error(y_test,y_pred))
```

These values indicate the performance of a regression model. Lower values for both MSE and MAE indicate better performance, as they represent smaller errors between predicted and actual values.
